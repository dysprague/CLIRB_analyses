## Fine tuning SLEAP/ DANNCE models

1. select frames to be used for fine tune training
2. take current 3D outputs and unproject into 2D camera spaces to produce n_frames x n_cameras x 23 x 3 
3. convert unprojected 2D keypoints into Label3D.saveState() mat format 
4. Label3D.loadState(file) load prelabeled points into Label3D, correct labels, and resave in DANNCE training format
5. Fine tune DANNCE model and rerun inference on existing sessions
6. Verify DANNCE inference outputs
7. Use DANNCE inference outputs to retrain SLEAP
